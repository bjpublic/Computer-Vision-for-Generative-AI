```python
import os
from solver import Solver
from data_loader import get_loader
from torch.backends import cudnn
from pathlib import Path

opt={
    'mode':'train',
    'lambda_cls':1,
    'lambda_rec':10,
    'lambda_gp':10,
    'dataset':'CelebA',
    'selected_attrs': ['Black_Hair', 'Blond_Hair', 'Brown_Hair', 
                       'Male', 'Young'],
    'batch':16,
    'num_iter':200000,
    'decay_iter':100000,
    'test_iter':200000,
    'celeba_path':'data/celeba/images',
    'attr_path':'data/celeba/list_attr_celeba.txt',
    'log_dir': 'stargan/logs',
    'sample_dir':'stargan/samples',
    'model_save_dir':'stargan/models',
    'result_dir':'stargan_test/results',
}
```

```python
# StarGAN 학습을 위한 셀입니다. 사전 학습 모델을 사용한 실습을 수행하려면 다음 셀로 넘어가세요.

cudnn.benchmark = True

Path(opt['log_dir']).mkdir(exist_ok=True, parents=True)
Path(opt['model_save_dir']).mkdir(exist_ok=True, parents=True)
Path(opt['sample_dir']).mkdir(exist_ok=True, parents=True)
                 
celeba_loader = None
rafd_loader = None

if opt['dataset'] in ['CelebA']:
    celeba_loader = get_loader(opt['celeba_path'], opt['attr_path'], 
                               opt['selected_attrs'], 178, 128, opt['batch'],
                               'CelebA', opt['mode'])

solver = Solver(celeba_loader, rafd_loader, opt)

if opt['mode'] == 'train':
    solver.train()
```

```python
# StarGAN 영상 변환을 위한 셀입니다. 

cudnn.benchmark = True

opt['mode'] = 'test'
opt['model_save_dir'] = 'stargan_celeba_128/models'
opt['celeba_path'] = 'data/celeba/images'
opt['test_num'] = 10
Path(opt['result_dir']).mkdir(exist_ok=True, parents=True)

celeba_loader = None
rafd_loader = None

if opt['dataset'] in ['CelebA']:
    celeba_loader = get_loader(opt['celeba_path'], opt['attr_path'],
                               opt['selected_attrs'], 178, 128, 
                               opt['batch'], 'CelebA', opt['mode'])

solver = Solver(celeba_loader, rafd_loader, opt)

if opt['mode'] == 'test':
    solver.test()
```
